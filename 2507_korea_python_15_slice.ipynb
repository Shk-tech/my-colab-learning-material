```python
# 1. 퍼즐 랜덤 생성 0-14, -1
# while True:
#    0. 현재 상태 출력
#    1. 입력
#    2. 입력대로 움직이기
#    3. 퍼즐이 맞았는지 확인.
```

```python
# 퍼즐 랜덤생성
# 0~ 14 까지 숫자를 셔플
board = [[-1]*4 for _ in range(4)]
def shuffle_board():
    import random
    data = list(range(0, 15))
    random.shuffle(data)
    data.append(-1)

    i = 0
    for y in range(4):
        for x in range(4):
            board[y][x] = data[i]
            i += 1
```

```python
def print_board():
    for y in range(4) :
        for x in range(4) :
            num = board[y][x]
            if num >= 0:
                print(f'{board[y][x]:2X}', end='  ')
            else:
                print(f' □', end='  ')
        print()
shuffle_board()
print_board()
```

```python
def input_move():
    text = input('>>')
    if text == 'w':
        dx = 0
        dy = 1
    if text == 'a':
        dx = 1
        dy = 0
    if text == 's':
        dx = 0
        dy = -1
    if text == 'd':
        dx = -1
        dy = 0

    # -1블럭 찾기
    bx = -1
    for y in range(4):
        for x in range(4):
            if board[y][x] == -1:
                bx = x
                by = y
                break
        if bx != -1:
            break

    tx = bx + dx
    ty = by + dy

    if tx < 0 or tx > 3 or ty < 0 or ty > 3:
        return

    # board[by][bx], board[ty][tx] = board[ty][tx], board[by][bx]
    board[by][bx] = board[ty][tx]
    board[ty][tx] = -1
```

```python
import datetime
print(datetime.datetime.now())
shuffle_board()
while True:
    print_board()
    input_move()
```
