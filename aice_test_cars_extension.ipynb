# 1
Pandas 함수로 2개 데이터 파일을 읽고 합쳐서 1개의 데이터프레임 변수명 df에 할당하는 코드를 작성하세요.
 - cars.csv파일을 읽어 데이터프레임 변수명 df_a에 할당하세요.
 - cars_extension.json 파일을 읽어 데이터 프레임 변수명 df_b에 할당하세요.
 - df_a 뒤에 df_b 데이터프레임을 합치세요. concat 함수를 활용하여 총 100행의 데이터를 완성하세요.
    - 합치는 축 방향 (axis) : 0
    - 합치는 방법(join) : 'outer'
    - 인덱스 보존(ignore_index) : True


```python
import pandas as pd
df_a = pd.read_csv('cars.csv')
df_b = pd.read_json('cars_extension.json')
df = pd.concat([df_a, df_b], join='outer', ignore_index=True, axis=0)
df
```
