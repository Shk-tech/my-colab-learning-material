```python
# 머신러닝의 조건
# 1. 정량적 성능 측정
# 2. 자동으로 성능을 향상시키는 방법
#   -> 해석적방법 (수식)
#   -> 경사하강법 (딥러닝)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# artificial 샘플
m = 100
# x : 0~2
# y : x * 3 + 4
# x = []
# y = []
# for _ in range(m):
#     xi = np.random.random() * 2
#     yi = xi * 3 + 4 + np.random.randn()
#     x.append(xi)
#     y.append(yi)
x = np.random.rand(m, 1) * 2
y = x * 3 + 4 + np.random.randn(m, 1)
plt.plot(x, y, '.')
```

```python
ones = np.ones((m, 1))
xb = np.concat([x, ones], axis=1)
xbp = np.linalg.inv(xb.T @ xb) @ xb.T
theta = xbp @ y
theta
```

```python
ea = theta[0, 0]
eb = theta[1, 0]
def model(x):
    return x* ea + eb

tx = np.array([0, 2])
ty = model(tx)
plt.plot(tx, ty, '-')
plt.plot(x, y, '.')
```

# 2차 다항식

```python
# artificial
m = 100
x = np.random.rand(m, 1) * 6 - 3  #-3 ~ +3
y = 0.5 * x**2 + x + 2 + np.random.randn(m, 1)
plt.plot(x, y, '.')
```

```python
xb = np.c_[x**2, x**1, x**0]
theta = np.linalg.pinv(xb) @ y
theta

# x : (N, 1) 입력
# y : (N, 1) 출력
def model(x):
    return np.c_[x**2, x**1, x**0] @ theta

#tx = np.array([-3, -2, -1, 0, 1, 2, 3]).reshape(-1, 1)
tx = np.linspace(-3, 3, 100).reshape(-1, 1)
ty = model(tx)
plt.plot(tx, ty, '-')
plt.plot(x, y, '.')
```
