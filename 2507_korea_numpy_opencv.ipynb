```python
# OpenCV
# Open Computer Vision libray
# 영상 - 이미지, 동영상 - 비디오
# Intel...! 1999 영상처리 붐!
# 영상라이브러리의 표준 (C, C++)
# Python (OpenCV, PIL)
# OpenCV는 image를 위한 클래스를 따로 만들지 않고
# numpy행렬을 그냥 차용함! -> 넘파이 연산을 바로 영상처리로 활용
```

```python
# 이미지 데이터에 대한 이해
# 하나의 픽셀 uint8 - 0~255
# 인공지능, 특별한 경우 float32 0~1
import numpy as np
import matplotlib.pyplot as plt # matplitlib -> Matlab에 있는 plot 라이브러리 카피캣!!
import cv2 # opencv
```

```python
a = np.zeros((5, 5, 3), np.uint8)
# y, x, c   HWC -> plt.imshow가 행렬을 HWC, RGB로 해석
a[0, 0, 0] = 255
a[1, 2, 1] = 255
a[2, :, 2] = 255
a[:, 4] = [70, 130, 55]
a[3:, :3] = [255, 250, 252]
plt.imshow(a)
```

```python
cv2.imwrite('test.png', a)
```
