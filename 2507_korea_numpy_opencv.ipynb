```python
#numpy H, W...
#opencv W, H...
# cv2.resize(image, (W//2, H//2))
src = cv2.resize(image, None, fx=0.6, fy=0.6)
src
```

```python
# brightness(+-), constrast(*)
b = np.array([10, 0, -5]) #white balance
c = np.array([0.8, 1.2, 1.])
f = src.astype(np.float32)
n = np.random.randn(*src.shape) * 10
f = f * c + ((1 - c) * 127 + b) + n
f[f > 255] = 255
f[f < 0] = 0
dst = f.astype(np.uint8)
dst
```

```python
# 1. 개별 학생들의 총점을 구하시오.
a.sum(axis=1)
```

```python
# 2. 수학 점수의 평균을 구하시오.
# 2.1 개별 과목의 평균(mean)
a.mean(axis=0)[0]

# 2.2 수학점수만 먼저 뽑아내고 -> mean
a[:, :1].mean()
```

```python
# 3. 수학점수가 높으면 과학 점수도 높은가?
# 수학점수가 60 이상인 학생들의 과학 점수 평균 vs
# 수학점수가 60 미만인 학생들의 과학 점수 평균
a[a[:, 0] >= 60, :].mean(axis=0)[1]
cond = (a[:, 0] >= 60)
a[cond, 1].mean()
a[~cond, 1].mean()
```
