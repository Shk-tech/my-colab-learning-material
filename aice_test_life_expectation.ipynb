# 1
Pandas 함수로 2개 데이터 파일을 읽고 합쳐서 1개의 데이터프레임 변수명 df에 할당하는 코드를 작성하세요.
 - LifeExpA.csv파일을 읽어 데이터프레임 변수명 df_a에 할당하세요.
 - LifeExpB 파일을 읽어 데이터 프레임 변수명 df_b에 할당하세요.
 - df_a, df_b를 합쳐서 하나의 데이터프레임으로 만드세요. 판다스의 concat을 활용하세요

합친 데이터셋의 index를 초기화하세요
 - reset_index 활용
 - 기존 인덱스 버림(drop) : True

처리 결과를 df에 저장하세요.

```python
' asdf '.strip()
```

```python
import pandas as pd
df_a = pd.read_csv('LifeExpA.csv')
df_b = pd.read_csv('LifeExpB.csv')
df = pd.concat([df_a, df_b])
df = df.reset_index(drop=True)
df.columns = df.columns.str.strip() #문자열 취급하면서 elementwise
df.columns
```

# 1
가이드에 따라 원-핫 인코딩을 수행합니다. pandas의 get_dummies 함수를 활용하세요.
 - 대상 : df3의 object 타입의 열
 - 전처리가 반영된 결과를 df4에 저장하세요.

```python
df3['Polio'].replace({
    'very low':1,
    'low':2,
    'medium':3,
    'high':4,
    'very high':5
})
```

```python
df4 = pd.get_dummies(df3)
df4
```

# 1
 - 데이터를 입력 X와 타깃 y로 분리하시오.
   - y : Life expectancy
   - X : Life expectancy 외 컬럼
 - sklearn의 train_test_split 함수를 활용하여 학습데이터와 테스트 데이터로 분리하시오.
   - random_state : 42
   - 학습데이터 80%, 훈련데이터 20%

# 1
 - RobustScaler를 활용하여 입력 데이터 X_train, X_test에 대한 스케일링을 진행하시오.
   - X_train : fit_transform 활용
   - X_test : transform 활용


```python
from sklearn.model_selection import train_test_split
y = df4['Life expectancy']
x = df4.drop('Life expectancy', axis=1)
test_size = int(len(df4) * 0.2)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=test_size, random_state=42)
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
# scaler.fit(x_train)
# x_train = scaler.transform(x_train)
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

#

```python
# fit(x, y) 동시에 받아서 fit -> predict
# fit(x)                    -> transform
```

```python
from sklearn.linear_model import *
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
# train sample mse
pred = lin_reg.predict(x_train)
# ((pred - y_train) ** 2).mean()
lin_train_mse = mean_squared_error(y_train, pred)
lin_train_mse
```

```python
pred = lin_reg.predict(x_test)
lin_test_mse = mean_squared_error(y_test, pred)
lin_test_mse
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor()
rf_reg.fit(x_train, y_train)
pred = rf_reg.predict(x_train)
rf_train_mse = mean_squared_error(y_train, pred)

pred = rf_reg.predict(x_test)
rf_test_mse = mean_squared_error(y_test, pred)
rf_train_mse, rf_test_mse
```

```python
from sklearn.tree import DecisionTreeRegressor

dt_reg = DecisionTreeRegressor()
dt_reg.fit(x_train, y_train)
pred = dt_reg.predict(x_train)
dt_train_mse = mean_squared_error(y_train, pred)

pred = dt_reg.predict(x_test)
dt_test_mse = mean_squared_error(y_test, pred)
dt_train_mse, dt_test_mse
```
