```python
#filter 여러 가지 상황... 지금은 noise filter
#IMU (자력 + 가속도 + 중력)
#GPS (차량 속도 측정)

import numpy as np
import matplotlib.pyplot as plt
```

```python
# frequency, amplitude ->
t = np.linspace(0, 2, 100)
z = np.sin(1 * t * 2 * np.pi)
plt.plot(t, z)
```

```python
n = np.random.randn(100) * 0.5
plt.plot(t, z + n)
```

```python
# 가장 이상적인 필터
# x(t) = z(t) + n(t)
# x'(t) = (x(t-k)...x(t+k))/m
# 쓸 수 없는 이유... 실시간성
# 미래를 알 수 없기 때문에... t+? 확인.
filter = np.ones(9) / 9
z1 = np.convolve(z + n, filter)[4:-4]
plt.plot(t, z + n)
plt.plot(t, z1)
```

```python
filter = np.ones(9)
filter[-4:] = 0
filter = filter / 5
z2 = np.convolve(z + n, filter)[4:-4]
plt.plot(t, z + n)
plt.plot(t, z1, 'r')
plt.plot(t, z2, 'm')
```
